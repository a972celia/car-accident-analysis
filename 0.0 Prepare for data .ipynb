```python
import glob
import pandas as pd
```

```python
path=r'/Users/Celia/Desktop/Data_Assurance_Data_Test_S/Data_Dictionary'
all_files=glob.glob(path + "/*.csv")

dictionay = {}
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    dictionay[filename.split("/")[-1]]=df
```

```python
# Reference data codes
Ur = dictionay['Urban_Rural.csv'].set_index('code')
accident_data['Urban_or_Rural_Area'] = accident_data['Urban_or_Rural_Area'].apply(lambda x: Ur.loc[x])

LC = dictionay['Light_Conditions.csv'].set_index('code')
accident_data['Light_Conditions'] = accident_data['Light_Conditions'].apply(lambda x: LC.loc[x])

# LocAuthH = dictionay['Local_Authority_Highway.csv'].set_index('code')
# accident_data['Local_Authority_(Highway)'] = accident_data['Local_Authority_(Highway)'].apply(lambda x: LocAuthH.loc[x])

RC1 = dictionay['Road_Class.csv'].set_index('code')
accident_data['1st_Road_Class'] = accident_data['1st_Road_Class'].apply(lambda x: RC1.loc[x])

accident_data['2nd_Road_Class'].replace(-1,0,inplace=True) # Data does not match supplied data dictionary
#RC2 = dictionay['2nd Road Class'].set_index('code')
accident_data['2nd_Road_Class'] = accident_data['2nd_Road_Class'].apply(lambda x: RC1.loc[x])

pA = dictionay['Police_Officer_Attend.csv'].set_index('code')
accident_data['Police_Officer_Attend'] = accident_data['Police_Officer_Attend'].apply(lambda x: pA.loc[x])

JD = dictionay['Junction_Detail.csv'].set_index('code')
accident_data['Junction_Detail'] = accident_data['Junction_Detail'].apply(lambda x: JD.loc[x])

LocAuthD = dictionay['Local_Authority_District.csv'].set_index('code')
accident_data['Local_Authority_(District)'] = accident_data['Local_Authority_(District)'].apply(lambda x: LocAuthD.loc[x])

DoW = dictionay['Day_of_Week.csv'].set_index('code')
accident_data['Day_of_Week'] = accident_data['Day_of_Week'].apply(lambda x: DoW.loc[x])


JC = dictionay['Junction_Control.csv'].set_index('code')
accident_data['Junction_Control'] = accident_data['Junction_Control'].apply(lambda x: JC.loc[x])

PCPh = dictionay['Ped_Cross_Physical.csv'].set_index('code')
accident_data['Ped_Cross_Physical'] = accident_data['Ped_Cross_Physical'].apply(lambda x: PCPh.loc[x])

accSev = dictionay['Accident_Severity.csv'].set_index('code')
accident_data['Accident_Severity'] = accident_data['Accident_Severity'].apply(lambda x: accSev.loc[x])

RT = dictionay['Road_Type.csv'].set_index('code')
accident_data['Road_Type'] = accident_data['Road_Type'].apply(lambda x: RT.loc[x])

PCH = dictionay['Ped_Cross_Human.csv'].set_index('code')
accident_data['Ped_Cross_Human'] = accident_data['Ped_Cross_Human'].apply(lambda x: PCH.loc[x])

pf = dictionay['Police_Force.csv'].set_index('code')
accident_data['Police_Force'] = accident_data['Police_Force'].apply(lambda x: pf.loc[x])










```
